```python
user_input = """ Good morning, how are you """
```

```python
import time
import tensorflow as tf

one_step = tf.saved_model.load(r'C:\Users\chamo\Documents\GitHub\Cham-GPT\Cham-GPT\saved_models\Cham-GPT model')

start = time.time()
states = None
next_char = tf.constant(['Jas:\n'+str(user_input)+'\n\nCham:\n'])
result = [next_char]

for n in range(100):
  next_char, states = one_step.generate_one_step(next_char, states=states)
  result.append(next_char)

result = tf.strings.join(result)
end = time.time()
print(result[0].numpy().decode('utf-8'), '\n\n' + '_'*80)
print('\nRun time:', end - start)
```
